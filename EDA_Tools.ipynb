# **Matplotlib**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# **Types of Basic Plots in Matplotlib**

```python
# Line Plot – shows trends over time
# Bar Chart – compares different categories
# Histogram – shows data distribution
# Scatter Plot – shows relationships between two variables
import matplotlib.pyplot as plt

products = ["A","B","C"]
sales = [100,150,200]

plt.bar(products, sales)
plt.title("Product Sales")
plt.show()
```

# **Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.show()
```

# **Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Create heatmap
sns.heatmap(numeric_data.corr(), annot=True)

plt.show()
```

# **Pairplot**

```python
sns.pairplot(data)
plt.show()
```

# ***Hands-On Programs***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Sales":[100,200,150,300,250],
    "Profit":[20,40,30,60,50]
}

df = pd.DataFrame(data)

plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

# ***Client Project***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Sales":[100,200,150,300,250],
    "Profit":[20,40,30,60,50],
    "Region":["East","West","East","North","West"]
}

df = pd.DataFrame(data)

# Scatter Plot
plt.figure()
sns.scatterplot(x="Sales", y="Profit", data=df)
plt.title("Sales vs Profit")

# Histogram
plt.figure()
plt.hist(df["Sales"])
plt.title("Sales Distribution")

# Bar Chart
plt.figure()
sns.barplot(x="Region", y="Sales", data=df)
plt.title("Sales by Region")

plt.show()
```
